```scala
implicit class EnhancesList(students: List[student]) {
    def %>(value: Int): List[student] = {
        var newlist: List[student] = List()
        for(i <- students){
            if(i.score > value){
                newlist = i :: newlist
            }
        }
        return newlist
    }

    def %<(value: Int): List[student] = {
        var newlist2: List[student] =  List()
        for(i <- students){
            if(i.score < value){
                newlist2 = i :: newlist2
            }
        }
        return newlist2
        
    }
}

```
