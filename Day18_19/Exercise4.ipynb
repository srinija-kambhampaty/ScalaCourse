```scala
import org.apache.spark.sql.SparkSession
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types.{StructType, StructField, StringType, DoubleType}
import org.apache.spark.sql.Row
import scala.util.Random

val spark = SparkSession.builder
  .appName("Exercise 4")
  .master("yarn")
  .getOrCreate()

def randomElement[T](seq: Seq[T]): T = {
    val random = new Random()
    seq(random.nextInt(seq.length))
  }
```

```scala
val data = Seq(
      (1, "completed", 100.5, 101, "2023-01-01 01:00:00"),
      (2, "pending", 200.0, 102, "2023-01-01 01:01:00"),
      (3, "completed", 150.25, 103, "2023-01-01 01:02:00"),
      (4, "canceled", 300.0, 104, "2023-01-01 01:03:00"),
      (5, "completed", 250.75, 105, "2023-01-01 01:04:00")
    )

val schema = Seq("order_id", "status", "amount", "customer_id", "order_date")
```

```scala
 val df = spark.createDataFrame(data).toDF(schema: _*)
 df.show()
```

```scala
val filePath = "gs://srinija/sample_data/data.parquet" 
df.write.parquet(filePath)
println(s"Parquet file created at: $filePath")
```

```scala
val inputPath = "gs://srinija/sample_data/data.parquet" 
val outputPath = "gs://srinija/output/processed_data.parquet" 
```

```scala
val inputData = spark.read.parquet(inputPath)
val processedData = inputData.filter($"status" === "completed")
```

```scala
 processedData.write
      .mode("overwrite")
      .parquet(outputPath)
println("Processing completed and data written to GCS.")
```

```scala
spark.stop()
```
